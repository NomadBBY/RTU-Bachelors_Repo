```python
import numpy as np
import matplotlib.pyplot as plt

def create_spectrogram(data, sampling_rate, window_size=256, overlap=128):
    """
    Create a spectrogram from audio data.

    Parameters:
        data (ndarray): Audio data.
        sampling_rate (int): Sampling rate of the audio data.
        window_size (int): Size of the FFT window.
        overlap (int): Overlap between consecutive windows.

    Returns:
        ndarray: Spectrogram data.
        ndarray: Frequencies.
        ndarray: Time intervals.
    """
    num_samples = len(data)
    num_overlap = window_size - overlap
    num_windows = (num_samples - window_size) // num_overlap + 1

    frequencies, time_intervals, spectrogram = \
        plt.specgram(data, NFFT=window_size, Fs=sampling_rate,
                     noverlap=overlap, scale='dB')

    return spectrogram, frequencies, time_intervals
```
